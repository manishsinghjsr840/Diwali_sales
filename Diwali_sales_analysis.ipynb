# Exploratory Data Analysis

## Gender

```python
sns.countplot(x="Gender",data=df)
```

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
     ax.bar_label(bars)
    
```

```python
sales_gen=df.groupby(["Gender"], as_index= False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen
```

```python
ay=sns.barplot(x="Gender",y="Amount",data=sales_gen)
```

here we can see that most of the buyers are female and even purchasing power of female is more than male


# AGE Group



```python
 ag=sns.countplot(x="Age Group",data=df,hue="Gender")
```

```python
 ag=sns.countplot(x="Age Group",data=df,hue="Gender")
    
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
sale_age_group=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="Age Group",y="Amount",data=sale_age_group)
```

From above graph we can see that most of the buyers is from age group 26-35 female
